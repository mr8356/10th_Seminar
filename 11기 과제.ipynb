```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
study_performace = pd.read_csv('./study_performance.csv')
```

```python
# sns.histplot(data)
df = pd.DataFrame(study_performace)
# category에 따라 다른 색으로 히스토그램 그리기
sns.histplot(data=df['math_score'])
plt.show()
```

```python
sns.boxplot(data=df[['math_score','reading_score','writing_score']])
plt.show()
```

```python
sns.violinplot(data=df[['math_score','reading_score','writing_score']])
plt.show()
```

```python
sns.countplot(x='gender', data=df[['gender','math_score']])
plt.show()
```
